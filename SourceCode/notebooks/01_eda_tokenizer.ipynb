```python
from transformers import AutoTokenizer
import pandas as pd

# 1. Load Tokenizer
model_name = "VietAI/vit5-base"
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)

# 2. Load thử một vài dòng dữ liệu lỗi bạn đã tạo
df = pd.read_csv(r"C:\Users\ASUS\Documents\GitHub\THUC-TAP-CO-SO\SourceCode\data\processed\train_data.csv")
sample_noisy_text = df.iloc[0]['source'] 
print(f"Câu lỗi gốc: {sample_noisy_text}")

# 3. Xem cách Tokenizer chia từ
tokens = tokenizer.tokenize(sample_noisy_text)
ids = tokenizer.convert_tokens_to_ids(tokens)

print(f"Tokens: {tokens}")
print(f"Token IDs: {ids}")
```
